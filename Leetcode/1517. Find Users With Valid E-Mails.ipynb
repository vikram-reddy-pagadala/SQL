# ***1517. Find Users With Valid E-Mails***
``` console
Table: Users
+---------------+---------+
| Column Name   | Type    |
+---------------+---------+
| user_id       | int     |
| name          | varchar |
| mail          | varchar |
+---------------+---------+

user_id is the primary key (column with unique values) for this table.
This table contains information of the users signed up in a website. Some e-mails are invalid.
 

Write a solution to find the users who have valid emails.

A valid e-mail has a prefix name and a domain where:

The prefix name is a string that may contain letters (upper or lower case), digits, underscore '_', period '.', and/or dash '-'. The prefix name must start with a letter.
The domain is '@leetcode.com'.
Return the result table in any order.

The result format is in the following example.

 

Example 1:

Input: 
Users table:
+---------+-----------+-------------------------+
| user_id | name      | mail                    |
+---------+-----------+-------------------------+
| 1       | Winston   | winston@leetcode.com    |
| 2       | Jonathan  | jonathanisgreat         |
| 3       | Annabelle | bella-@leetcode.com     |
| 4       | Sally     | sally.come@leetcode.com |
| 5       | Marwan    | quarz#2020@leetcode.com |
| 6       | David     | david69@gmail.com       |
| 7       | Shapiro   | .shapo@leetcode.com     |
+---------+-----------+-------------------------+

Output: 
+---------+-----------+-------------------------+
| user_id | name      | mail                    |
+---------+-----------+-------------------------+
| 1       | Winston   | winston@leetcode.com    |
| 3       | Annabelle | bella-@leetcode.com     |
| 4       | Sally     | sally.come@leetcode.com |
+---------+-----------+-------------------------+

Explanation: 
The mail of user 2 does not have a domain.
The mail of user 5 has the # sign which is not allowed.
The mail of user 6 does not have the leetcode domain.
The mail of user 7 starts with a period.
```

**Solutions:**

`````python
SELECT * FROM users WHERE mail REGEXP '^[a-zA-Z]+[a-zA-Z0-9._-]*@leetcode\\.com$';
`````

``` console
In SQL, regular expressions (regex) allow you to search for patterns within strings. MySQL supports regular expressions through the REGEXP or RLIKE operators.

Syntax:
column_name REGEXP 'pattern'

Example:

SELECT * FROM table_name WHERE column_name REGEXP 'pattern';

Common Regex Patterns:
^: Matches the start of a string.
$: Matches the end of a string.
.: Matches any single character.
[abc]: Matches any character inside the brackets.
[a-z]: Matches any character in the range from a to z.
|: Acts as OR (alternation).
*: Matches 0 or more occurrences of the preceding element.
+: Matches 1 or more occurrences of the preceding element.
{n,m}: Matches between n and m occurrences of the preceding element.

Example Queries:
Match rows where name starts with 'A':

SELECT * FROM employees WHERE name REGEXP '^A';
Match rows where email ends with '.com':


SELECT * FROM users WHERE email REGEXP '.com$';
Match rows where phone_number contains exactly 3 digits followed by a hyphen and 4 digits:


SELECT * FROM contacts WHERE phone_number REGEXP '^[0-9]{3}-[0-9]{4}$';
Match rows where address contains the word "Street" or "Avenue":


SELECT * FROM addresses WHERE address REGEXP 'Street|Avenue';
Match rows where username contains only letters and numbers (no special characters):


SELECT * FROM users WHERE username REGEXP '^[a-zA-Z0-9]+$';
Case-Sensitivity:
By default, REGEXP is case-insensitive. To make it case-sensitive, use the BINARY keyword:

SELECT * FROM table_name WHERE BINARY column_name REGEXP 'pattern';
Negation (NOT REGEXP):
You can negate a regex match using NOT REGEXP:


SELECT * FROM users WHERE username NOT REGEXP '^[a-zA-Z0-9]+$';

Using Regex in Other Databases:
PostgreSQL: Use ~ or ~* (case-insensitive).
SQL Server: Use PATINDEX or LIKE (regex is not natively supported, but similar functionality exists).
Oracle: Use REGEXP_LIKE().
```

``` console
REGEXP '^[a-zA-Z0-9._%+-]+@leetcode\.com$'
Explanation:
^[a-zA-Z0-9._%+-]+: Matches the local part of the email (before the @ symbol). It includes:

a-zA-Z0-9: Alphanumeric characters.
._%+-: Special characters allowed in email addresses.
+: Ensures one or more of these characters.
@leetcode\.com: Matches exactly @leetcode.com.

Note: The . is escaped as \. because . is a special regex character that matches any single character. To match an actual period, you need to escape it.
$: Ensures the string ends after @leetcode.com.
```

``` console
If you want the email to start with letters but still allow numbers, dots (.), underscores (_), or hyphens (-) after the initial letters, you can modify your regular expression as follows:

Correct Regular Expression:

SELECT * FROM users WHERE mail REGEXP '^[a-zA-Z]+[a-zA-Z0-9._-]*@leetcode\.com$';
Explanation:
^[a-zA-Z]+:

^: Ensures the string starts at the beginning.
[a-zA-Z]+: The first part of the email must start with one or more letters (uppercase or lowercase).
[a-zA-Z0-9._-]*:

This allows any combination of letters, numbers, dots, underscores, or hyphens after the initial letters.
* means zero or more occurrences, so the local part can consist only of letters or can be followed by these characters.
@leetcode\.com:

Matches the domain @leetcode.com.
The period (.) is escaped as \. to match a literal dot.
$: Ensures the string ends after @leetcode.com.

Final Query:

SELECT * FROM users WHERE mail REGEXP '^[a-zA-Z]+[a-zA-Z0-9._-]*@leetcode\.com$';
```

`````python
SELECT *
FROM Users
WHERE mail REGEXP '^[A-Za-z][A-Za-z0-9_\.\-]*@leetcode(\\?com)?\\.com$';
`````

``` console
Regex Pattern: ^[A-Za-z][A-Za-z0-9_\.\-]*@leetcode(\?com)?\.com$

Let's break down the regex pattern step by step to clearly explain each part:

^: Anchor the regex pattern to match from the start of the string.

[A-Za-z]: Match any single uppercase or lowercase letter. The email prefix name must start with a letter.

[A-Za-z0-9_.-]*: Match any number of characters following the first letter in the email prefix name. It includes letters (upper or lower case), digits, underscore '_', period '.', and/or dash '-'.

@: Match the literal '@' character, which separates the prefix name and the domain.

leetcode: Match the literal 'leetcode', which is part of the email domain.
(?com)?: Make the sequence ?com optional in the email domain. Allows the pattern to match both '@leetcode.com' and '@leetcode?com'.

. : Match the literal '.' character, which separates the 'leetcode' part from the 'com' part of the domain.
com: Match the literal 'com' at the end of the email domain.

$: Anchor the regex pattern to match until the end of the string.

Explanation:
The regex pattern ensures that a valid email must have the following format:

The prefix name starts with a letter.
The prefix name can contain letters (upper or lower case), digits, underscore '_', period '.', and/or dash '-'.
The domain must be '@leetcode.com' with an optional '?com' part.
````
