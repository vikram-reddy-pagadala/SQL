Q.Write an SQL query to fetch the runs scored in each over.

Expected Output:

+-------+------+
| OVERS | RUNS |
+-------+------+
|     1 |   11 |
|     2 |   14 |
|     3 |    1 |
+-------+------+

```

### **Solution-1:**

``` sql
SELECT ceil(ball_cnt/6) as OVERS, sum(run) as RUNS
FROM cricket
group by ceil(ball_cnt/6);
