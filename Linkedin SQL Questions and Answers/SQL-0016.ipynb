**Solution:**
``` sql
select *, 
sum(
round(
case 
when status = 'debit' and mode = 'Credit Card' then -amount * (100-2)/100 
when status = 'debit' then -amount
else amount end,0)
)
over(partition by custid order by txndate, mode asc)  as net_balance
from bankstatement;
```
